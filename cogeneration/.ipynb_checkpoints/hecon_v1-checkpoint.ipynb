# <span style="color:#2E86C1;">**Pinch Analysis: Maximizing Energy Efficiency**</span>

Enhancing **energy efficiency** and **heat integration** is paramount for industries aiming to reduce utility costs and environmental impact. **Pinch Analysis** provides a **systematic** framework to optimize heat exchange among **hot** and **cold** streams, ultimately minimizing the need for external utilities.

---

## <span style="color:#8E44AD;">**Why Pinch Analysis Matters**</span>

- ‚≠ê **Lower Utility Costs:** Through targeted **heat recovery**, processes will rely less on external heating/cooling, reducing fuel and operational expenses.  
- ‚≠ê **Reduced Emissions:** By curbing external energy usage, we naturally cut down on **greenhouse gas** emissions.  
- ‚≠ê **Process Insight:** Pinch Analysis exposes critical **temperature intervals** (pinch points) where improvements in **heat exchange** are most impactful.

---

## <span style="color:#D35400;">**Core Concepts**</span>

1. <span style="color:#2980B9;">**Hot vs. Cold Streams**</span>  
   - **Hot Stream:** Must be **cooled** from $T_{\text{supply}}$ to $T_{\text{target}}$.  
   - **Cold Stream:** Must be **heated** from $T_{\text{supply}}$ to $T_{\text{target}}$.

2. <span style="color:#2980B9;">**Heat Capacity Flowrate**</span>  
   - Each stream has a **heat capacity**, $C_p$, in units like $\mathrm{J/(kg\cdot K)}$.  
   - The **heat load** (duty) can be computed as  
     $$ Q = \bigl(T_{\text{target}} - T_{\text{supply}}\bigr)\, C_p. $$

3. <span style="color:#2980B9;">**Minimum Temperature Approach, $\Delta T_{\min}$**</span>  
   - Ensuring **feasible** heat exchange demands a **minimum temperature difference** between hot and cold streams.  
   - **Pinch** procedures **shift** hot streams *down* and cold streams *up* by $\tfrac{\Delta T_{\min}}{2}$ before constructing composite curves or applying the **Problem Table Algorithm**.

4. <span style="color:#2980B9;">**Pinch Point**</span>  
   - The **pinch** is the temperature (in shifted coordinates) where the heat cascade reaches zero.  
   - It **separates** the network into regions above and below the pinch, dictating **where** external utilities (hot or cold) are strictly needed.

---

## <span style="color:#16A085;">**Composite Curves & Grand Composite Curve**</span>

- üîπ **Hot Composite Curve (HCC):** Illustrates **available heat** from all hot streams as temperature decreases.  
- üîπ **Cold Composite Curve (CCC):** Depicts the **required heat** by cold streams as temperature rises.  
- üîπ **Grand Composite Curve (GCC):** Formed by **subtracting** the cold curve from the hot curve over a **common temperature** grid. It highlights **net heat surplus or deficit** at every temperature level.

**Why use these curves?**  
They give a **visual** snapshot of the **thermal profile** of the process, showing **where** we can recover heat internally and where external utilities must fill in.

---

## <span style="color:#E74C3C;">**How This Notebook Helps**</span>

- Prompts you for **process data** (temperatures, $C_p$, or $Q$) in either $\mathrm{^\circ C}$ or K.  
- Converts everything into **SI** units internally:
  - $T$ in $\mathrm{^\circ C}$,  
  - $C_p$ in $\mathrm{J/(kg\cdot K)}$,  
  - $Q$ in $\mathrm{J}$.  
- **Pinch Analysis** (Problem Table Algorithm) finds:
  - **Pinch Temperature**  
  - **Minimum Hot Utility**  
  - **Minimum Cold Utility**  
  - **Heat Cascade** and **Interval Details**  
- **Composite Curves & GCC**: The code automatically:
  - Builds **Hot** and **Cold** Composite Curves,  
  - Merges them to form the **Grand Composite Curve**,  
  - **Plots** all three for easy interpretation.  
- **Optional Export**: You can **save** the interval table to CSV and **export** the final curves to a PNG (or any matplotlib-supported format).

**Outcome?**  
We will have a **quantitative** and **visual** results that pinpoint **energy-saving opportunities** in our process.

---

# <span style="color:#1ABC9C;">**Data Input & Automatic Unit Conversion for Pinch Analysis**</span>

The cell below guides you through **user-friendly** input prompts for:

- **Supply (Ts)** and **Target (Tt)** temperatures,
- **Heat capacity (Cp)** values,
- *(Optional)* **Heat duty (Q)** values.

---

## <span style="color:#9B59B6;">**Key Features**</span>

1. <span style="color:#3498DB;">**Temperature Units (Celsius or Kelvin)**</span>  
   - Prompts for Ts and Tt in either **¬∞C** or **K**.  
   - **Automatically** converts **Kelvin** inputs to **¬∞C** internally, ensuring consistent temperature differences and simplified calculations.

2. <span style="color:#3498DB;">**Heat Capacity (Cp) Units**</span>  
   - Accepts multiple unit options (**J/(kg¬∑K)**, **kJ/(kg¬∑K)**, **cal/(g¬∑K)**, **kcal/(kg¬∑K)**, **BTU/(lbm¬∑¬∞F)**).  
   - Converts them all to **J/(kg¬∑K)** behind the scenes.

3. <span style="color:#3498DB;">**Heat Duty (Q) Units**</span>  
   - If you provide **Q**, you can pick from **J**, **kJ**, **MJ**, **kWh**, **cal**, **kcal**, **BTU**, or **MWh**.  
   - The script normalizes your input to **J**.  
   - If you **omit** Q, it‚Äôs automatically computed via  
     $$Q = \bigl(T_{\text{target}} - T_{\text{supply}}\bigr) \times C_p.$$

4. <span style="color:#3498DB;">**SI-Conformant Internal Storage**</span>  
   - The final DataFrame columns are labeled:  
     - **Tsupply (¬∞C)**  
     - **Ttarget (¬∞C)**  
     - **Cp (J/(kg¬∑K))**  
     - **Q (J)**  
   - **Indices:** `'hot'` if $T_{\mathrm{supply}} > T_{\mathrm{target}}$; otherwise `'cold'`.

5. <span style="color:#3498DB;">**Reusable for Pinch Analysis**</span>  
   - Once confirmed, the **DataFrame** is fully in **SI units**, ready to be passed into pinch analysis or other **thermal calculations**.

---

## <span style="color:#F39C12;">**How to Use This Cell**</span>

1. **Run** the cell and respond to each **prompt** for data input.  
2. **Review** the printed DataFrame (in SI units).  
3. **Proceed** with pinch analysis or other steps, confident your units are **consistent** and **accurate**.


```python
# ------------------------------------------------------
# 1. Utility Dictionaries for Unit Conversion
# ------------------------------------------------------
# These dictionaries map a "choice" string (like '1','2', etc.)
# to a numeric multiplier, converting the user's unit to SI.

# Heat capacity (Cp) converted to J/(kg*K)
cp_unit_factors = {
    '1': 1.0,       # Already in J/(kg*K)
    '2': 1000.0,    # kJ/(kg*K) => J/(kg*K)  (1 kJ = 1000 J)
    '3': 4184.0,    # cal/(g*K) => J/(kg*K) (1 cal/g*K ~ 4.184 J/g*K => 4184 J/kg*K)
    '4': 4184.0,    # kcal/(kg*K) => J/(kg*K) (1 kcal/kg*K ~ 4184 J/kg*K)
    '5': 4186.8,    # BTU/(lbm*¬∞F) => J/(kg*K) (approx conversion)
}

# Labels to show in the prompts
cp_unit_labels = {
    '1': "J/(kg*K)",
    '2': "kJ/(kg*K)",
    '3': "cal/(g*K)",
    '4': "kcal/(kg*K)",
    '5': "BTU/(lbm*¬∞F)",
}

# Heat duty (Q) converted to Joules (J)
q_unit_factors = {
    '1': 1.0,       # Already in J
    '2': 1000.0,    # kJ => J (1 kJ = 1000 J)
    '3': 1e6,       # MJ => J (1 MJ = 1,000,000 J)
    '4': 3.6e6,     # kWh => J (1 kWh = 3.6e6 J)
    '5': 4.184,     # cal => J (1 cal ~ 4.184 J)
    '6': 4184.0,    # kcal => J (1 kcal = 4184 J)
    '7': 1055.06,   # BTU => J (approx. 1055.06 J in 1 BTU)
    '8': 3.6e9,     # MWh => J (1 MWh = 3.6e9 J)
}

# Labels to show in the prompts
q_unit_labels = {
    '1': "J",
    '2': "kJ",
    '3': "MJ",
    '4': "kWh",
    '5': "cal",
    '6': "kcal",
    '7': "BTU",
    '8': "MWh",
}

# ------------------------------------------------------
# 2. Parsing & Confirmation Utilities
# ------------------------------------------------------
def parse_input(input_str: str) -> List[float]:
    """
    Take a raw input string from the user, which may contain
    values separated by commas and/or spaces, and return
    a list of floats.

    Steps:
    1. Replace commas with spaces, so spaces separate everything.
    2. Attempt to convert each token into a float.
    3. If any token is non-numeric, raise ValueError.
    """
    cleaned_input = input_str.replace(',', ' ')  # unify delimiters as spaces
    try:
        return [float(num) for num in cleaned_input.split()]
    except ValueError:
        raise ValueError("Input contains non-numeric values, please try again.")

def confirm_input(data: List[float]) -> bool:
    """
    Show the user the parsed data for confirmation, asking
    "Is this correct? (yes/no)". If 'yes', returns True;
    otherwise returns False.

    This helps ensure the user didn't mistype values.
    """
    print(f"You entered: {data}")
    response = input("Is this correct? (yes/no): ").strip().lower()
    return response == 'yes'

def input_and_confirm(prompt: str, expected_length: Optional[int] = None) -> List[float]:
    """
    Continuously prompt the user with the given text, parse the input,
    and confirm it. Also optionally check that the parsed list
    has the expected length (e.g., if we already know how many
    data points we want).

    If the user denies correctness or the length is wrong,
    it loops until a valid, confirmed list of floats is obtained.
    """
    while True:
        user_input = input(prompt)
        try:
            parsed_data = parse_input(user_input)  # parse the string
            # If we have an expected number of values, enforce it:
            if expected_length is not None and len(parsed_data) != expected_length:
                print(f"Expected {expected_length} entries, but got {len(parsed_data)}. Please re-enter.\n")
                continue

            # Ask user to confirm
            if confirm_input(parsed_data):
                return parsed_data
            else:
                print("Let's try entering the data again.\n")

        except ValueError as e:
            # If parsing fails, we show the error and re-prompt
            print(str(e))
            print("Please re-enter the data correctly.\n")

# ------------------------------------------------------
# 3. Core Calculations
# ------------------------------------------------------
def calculate_heating_duties(ts_list: List[float],
                             tt_list: List[float],
                             cp_list: List[float]) -> List[float]:
    """
    Given supply temperatures (Ts), target temperatures (Tt),
    and heat capacities (Cp), compute the heat duties (Q).

    Formula:
        Q = (Ttarget - Tsupply) * Cp

    - ts_list, tt_list in ¬∞C
    - cp_list in J/(kg*K)
    - returns Q in Joules (J)
    """
    return [(tt - ts)*cp for ts, tt, cp in zip(ts_list, tt_list, cp_list)]

def calculate_heat_capacity(ts_list: List[float],
                            tt_list: List[float],
                            q_list: List[float]) -> List[float]:
    """
    Given supply temperatures (Ts), target temperatures (Tt),
    and known heat duties (Q), compute heat capacities (Cp).

    Formula:
        Cp = Q / (Ttarget - Tsupply)

    - ts_list, tt_list in ¬∞C
    - q_list in J
    - returns Cp in J/(kg*K)
    """
    cp_values = []
    for ts, tt, q_val in zip(ts_list, tt_list, q_list):
        deltaT = tt - ts
        if abs(deltaT) < 1e-9:
            # Avoid dividing by zero if Ttarget == Tsupply
            cp_values.append(0.0)
        else:
            cp_values.append(q_val / deltaT)
    return cp_values

# ------------------------------------------------------
# 4. Main Data Input Function
# ------------------------------------------------------
def source_data() -> pd.DataFrame:
    """
    The main function that orchestrates user inputs and conversions.

    Process Outline:
    A) Ask user if temperatures are in Celsius (C) or Kelvin (K).
       - If Kelvin, convert to ¬∞C internally by subtracting 273.15.
    B) Prompt user for supply (Ts) and target (Tt) temperatures,
       ensuring the lists match in length.
    C) Ask user which data they have for each stream: (1) Cp or (2) Q
       - If user has Cp, we parse/convert it to J/(kg*K) and then
         compute Q automatically.
       - If user has Q, we parse/convert it to J and then compute
         Cp automatically.
    D) Build a DataFrame of the final values (SI units):
         Tsupply (¬∞C), Ttarget (¬∞C), Cp (J/(kg*K)), Q (J)
       with an index: 'hot' if Tsupply > Ttarget, else 'cold'.
    E) Return this DataFrame for further use (e.g., pinch analysis).
    """
    # ------------------- A) Temperature unit -------------------
    while True:
        temp_unit = input("Are your supply/target temperatures in Celsius (C) or Kelvin (K)? ").strip().lower()
        if temp_unit in ("c", "k"):
            break
        else:
            print("Invalid choice. Please type 'C' or 'K'.\n")

    # ------------------- B) Ts & Tt Input ----------------------
    # 1) Supply Temperatures
    ts_list = input_and_confirm("Enter SUPPLY temperatures: ")
    if temp_unit == 'k':
        # Convert from Kelvin to ¬∞C
        ts_list = [v - 273.15 for v in ts_list]

    # 2) Target Temperatures
    tt_list = input_and_confirm(
        "Enter TARGET temperatures (must match # of supply temps): ",
        expected_length=len(ts_list)
    )
    if temp_unit == 'k':
        tt_list = [v - 273.15 for v in tt_list]

    # ------------------- C) Which data do we have (Cp or Q)? ---
    while True:
        print("\nWhich data do you have for each stream?")
        print("  1) Cp (we'll compute Q)")
        print("  2) Q (we'll compute Cp)")
        data_choice = input("Enter 1 or 2: ").strip()
        if data_choice in ('1','2'):
            break
        else:
            print("Invalid choice. Please enter 1 or 2.\n")

    # Initialize empty lists for cp_list and q_list
    cp_list = []
    q_list = []

    # --------- CASE 1: Cp only ----------
    if data_choice == '1':
        # 1) Prompt for the Cp unit
        print("\nChoose the unit for Cp:")
        for k, v in cp_unit_labels.items():
            print(f"  {k}) {v}")
        while True:
            cp_unit_choice = input("Enter your choice: ").strip()
            if cp_unit_choice in cp_unit_factors:
                break
            else:
                print("Invalid choice, please try again.\n")

        # 2) Prompt for Cp values, confirm length, convert to SI
        cp_vals = input_and_confirm(
            f"Enter Cp values in {cp_unit_labels[cp_unit_choice]} (matching # of streams): ",
            expected_length=len(ts_list)
        )
        factor_cp = cp_unit_factors[cp_unit_choice]
        cp_list = [c * factor_cp for c in cp_vals]

        # 3) Automatically compute Q
        q_list = calculate_heating_duties(ts_list, tt_list, cp_list)

    # --------- CASE 2: Q only ----------
    else:  # data_choice == '2'
        # 1) Prompt for the Q unit
        print("\nChoose the unit for Q (heat duty):")
        for k, v in q_unit_labels.items():
            print(f"  {k}) {v}")
        while True:
            q_unit_choice = input("Enter your choice: ").strip()
            if q_unit_choice in q_unit_factors:
                break
            else:
                print("Invalid choice, please try again.\n")

        # 2) Prompt for Q values, confirm length, convert to SI
        q_vals = input_and_confirm(
            f"Enter Q values in {q_unit_labels[q_unit_choice]} (matching # of streams): ",
            expected_length=len(ts_list)
        )
        factor_q = q_unit_factors[q_unit_choice]
        q_list = [qv * factor_q for qv in q_vals]

        # 3) Automatically compute Cp
        cp_list = calculate_heat_capacity(ts_list, tt_list, q_list)

    # ------------------- D) Generate Stream Labels -------------
    # If supply temp > target temp => 'hot', else => 'cold'
    label_list = ['hot' if ts > tt else 'cold' for ts, tt in zip(ts_list, tt_list)]

    # ------------------- E) Build & Return DataFrame -----------
    df = pd.DataFrame({
        'Tsupply (¬∞C)': ts_list,
        'Ttarget (¬∞C)': tt_list,
        'Cp (J/(kg*K))': cp_list,
        'Q (J)': q_list
    }, index=label_list)
    return df
```

# <span style="color:#34495E;">**Pinch Analysis Function**</span>

The `pinch_analysis` function below applies **Pinch Analysis**‚Äîa method to **minimize heating and cooling requirements** via the **Problem Table Algorithm**. It expects a **pandas DataFrame** containing:

- **"Tsupply (¬∞C)"**  
- **"Ttarget (¬∞C)"**  
- **"Cp (J/(kg*K))"**  

with rows labeled as **"hot"** or **"cold"** to distinguish streams.

---

## <span style="color:#2ECC71;">**How It Works**</span>

1. <span style="color:#2980B9;">**Shifting Temperatures**</span>  
   - Supply and target temperatures are **shifted** by $\pm \tfrac{\Delta T_{\min}}{2}$, based on whether the stream is **hot** or **cold**.  
   - **Hot** streams shift **down** by $\Delta T_{\min}/2$; **Cold** streams shift **up** by $\Delta T_{\min}/2$.

2. <span style="color:#2980B9;">**Building Temperature Intervals**</span>  
   - All **shifted** supply and target temperatures are gathered, **sorted (descending)**, and used to form **intervals** between each adjacent pair.

3. <span style="color:#2980B9;">**Net Heat Flow Across Intervals**</span>  
   - For each interval, we find a **net heat capacity** $C_{p,\text{net}}$:  
     - **Hot** streams contribute **positive** $C_p$.  
     - **Cold** streams contribute **negative** $C_p$.  
   - We multiply $C_{p,\text{net}}$ by the interval‚Äôs $\Delta T$ to get the **heat flow** in that segment.

4. <span style="color:#2980B9;">**Heat Cascade**</span>  
   - Starting from **0** at the highest shifted temperature, we **accumulate** each interval‚Äôs net heat flow from top to bottom.  
   - If this **cascade** goes below zero, we **shift** the entire profile **up** so its minimum is **zero**, removing negative heating artifacts.

5. <span style="color:#2980B9;">**Minimum Utilities**</span>  
   - After the shift, the **top** of the cascade reflects **minimum hot utility** (external heat).  
   - The **bottom** represents **minimum cold utility** (external cooling).

6. <span style="color:#2980B9;">**Pinch Temperature**</span>  
   - The **pinch** typically appears where the **shifted cascade** crosses **zero**.  
   - We detect this point and link it back to the corresponding **interval boundary** temperature.

7. <span style="color:#2980B9;">**Output**</span>  
   - Returns a **dictionary** containing:  
     - üî∏ **`pinch_temperature`**: Pinch point (shifted coordinate).  
     - üî∏ **`min_hot_utility`**, **`min_cold_utility`**: Required external energies (J).  
     - üî∏ **`intervals_df`**: DataFrame of intervals (net $C_p$, heat flows, cascade details).  
     - üî∏ **`cascade`**, **`shifted_cascade`**: The raw and adjusted cascade sequences.  
     - üî∏ **`intervals`**: The list of $(T_{\text{high}}, T_{\text{low}})$ interval boundaries.

Using this function provides a **systematic** way to identify **pinch** and utility targets‚Äîcore insights for **energy-saving** design and **heat integration**.


```python
def pinch_analysis(process_data: pd.DataFrame, dT_min: float = 10.0) -> dict:
    """
    Perform pinch analysis using the Problem Table Algorithm.

    Returns a dictionary with:
      - 'pinch_temperature'
      - 'min_hot_utility'
      - 'min_cold_utility'
      - 'intervals_df': DataFrame of interval details
      - 'cascade': list of unshifted cascade values
      - 'shifted_cascade': list of shifted cascade values
      - 'intervals': list of (T_high, T_low) tuples
    """

    # 1. Shift temperatures based on whether stream is hot or cold
    #    For hot streams: (Tshifted = Tsupply - ŒîTmin/2)
    #    For cold streams: (Tshifted = Tsupply + ŒîTmin/2)
    shifted_supply = []
    shifted_target = []

    for idx, row in process_data.iterrows():
        if idx.lower() == 'hot':
            # Hot stream: shift downward by dT_min/2
            shifted_supply.append(row['Tsupply (¬∞C)'] - dT_min/2)
            shifted_target.append(row['Ttarget (¬∞C)'] - dT_min/2)
        else:
            # Cold stream: shift upward by dT_min/2
            shifted_supply.append(row['Tsupply (¬∞C)'] + dT_min/2)
            shifted_target.append(row['Ttarget (¬∞C)'] + dT_min/2)

    # Store shifted temperatures in new columns
    process_data['Tsupply_shifted'] = shifted_supply
    process_data['Ttarget_shifted'] = shifted_target

    # 2. Collect all unique shifted temperatures and sort them descending
    all_temps = list(process_data['Tsupply_shifted']) + list(process_data['Ttarget_shifted'])
    unique_temps = sorted(set(all_temps), reverse=True)

    # Build intervals from each adjacent pair of unique temperatures
    intervals = [
        (unique_temps[i], unique_temps[i+1]) 
        for i in range(len(unique_temps) - 1)
    ]

    # 3. Calculate net heat flow (Cp_net * deltaT) in each interval
    interval_heat_flows = []
    cp_net_list = []  # We'll store net Cp for the interval

    for (T_high, T_low) in intervals:
        deltaT = T_high - T_low
        Cp_net = 0.0

        # Loop over each stream to see if it contributes in this interval
        for idx, row in process_data.iterrows():
            s_in = row['Tsupply_shifted']
            t_in = row['Ttarget_shifted']
            cp_value = row['Cp (J/(kg*K))']  # Updated column name for Cp

            # Check whether this interval lies fully within the stream's temperature range
            if (max(s_in, t_in) >= T_high) and (min(s_in, t_in) <= T_low):
                # If hot stream => add Cp
                # If cold stream => subtract Cp
                if idx.lower() == 'hot':
                    Cp_net += cp_value
                else:
                    Cp_net -= cp_value

        cp_net_list.append(Cp_net)
        interval_heat_flows.append(Cp_net * deltaT)

    # 4. Construct the heat cascade from top to bottom (unshifted)
    cascade = [0.0]
    for flow in interval_heat_flows:
        cascade.append(cascade[-1] + flow)

    # 5. Shift the cascade if any portion goes negative
    min_cascade_value = min(cascade)
    shift_amount = -min_cascade_value if min_cascade_value < 0 else 0.0
    shifted_cascade = [c + shift_amount for c in cascade]

    # 6. Minimum hot/cold utility after shifting
    # Top of cascade => min hot utility
    # Bottom of cascade => min cold utility
    min_hot_utility = shifted_cascade[0]
    min_cold_utility = shifted_cascade[-1]

    # 7. Pinch temperature => where the shifted cascade is zero
    pinch_temperature = None
    for i, val in enumerate(shifted_cascade):
        if abs(val) < 1e-9:
            if i < len(intervals):
                pinch_temperature = intervals[i][0]
            else:
                pinch_temperature = intervals[-1][1]
            break

    # 8. Create a neat DataFrame of intervals
    interval_rows = []
    for i, (T_high, T_low) in enumerate(intervals):
        row_dict = {
            'T_high': T_high,
            'T_low': T_low,
            'DeltaT': T_high - T_low,
            'Cp_net': cp_net_list[i],
            'Heat_flow': interval_heat_flows[i],
            'Cascade_in': cascade[i],
            'Cascade_out': cascade[i+1],
            'ShiftedCascade_in': shifted_cascade[i],
            'ShiftedCascade_out': shifted_cascade[i+1],
        }
        interval_rows.append(row_dict)

    intervals_df = pd.DataFrame(interval_rows)

    # Return all results in a dictionary
    return {
        'pinch_temperature': pinch_temperature,
        'min_hot_utility': min_hot_utility,
        'min_cold_utility': min_cold_utility,
        'intervals_df': intervals_df,
        'cascade': cascade,
        'shifted_cascade': shifted_cascade,
        'intervals': intervals,
    }
```

# <span style="color:#34495E;">**build_composite_curve Function**</span>

This function **generates a composite curve** (either **hot** or **cold**) from a process DataFrame containing:

- **"Tsupply (¬∞C)"** and **"Ttarget (¬∞C)"** (temperatures),
- **"Cp (J/(kg*K))"** (heat capacity),
- Row indices labeled **"hot"** or **"cold"**.

---

## <span style="color:#2ECC71;">**Steps**</span>

1. <span style="color:#2980B9;">**Filter by Stream Type**</span>  
   - If `stream_type='hot'`, we select rows labeled `'hot'`.  
   - Otherwise, we pick rows labeled `'cold'`.  

2. <span style="color:#2980B9;">**Shift Temperatures**</span>  
   - **Hot** streams shift **down** by $\Delta T_{\min}/2$.  
   - **Cold** streams shift **up** by $\Delta T_{\min}/2$.  
   - We store these shifts in `"Tsupply_shifted"` and `"Ttarget_shifted"` columns.

3. <span style="color:#2980B9;">**Unique Shifted Temperatures**</span>  
   - Collect all **shifted** supply and target temperatures.  
   - Sort them in **descending** order to form **intervals**.

4. <span style="color:#2980B9;">**Net Heat Capacity per Interval**</span>  
   - For each interval $(T_{\text{high}}, T_{\text{low}})$, sum up any $C_p$ from streams spanning that entire range.  
   - Multiply by the temperature difference $\Delta T$ to find the interval‚Äôs **heat** $\Delta Q$.

5. <span style="color:#2980B9;">**Cumulative Enthalpy**</span>  
   - Starting from **0** at the **top** (highest temperature), add each $\Delta Q$ in turn.  
   - This builds a **running enthalpy**‚Äîthe **composite curve**.

6. <span style="color:#2980B9;">**Output**</span>  
   - Returns a **DataFrame** with:
     - **"Temperature"**: The **boundary** temperatures (shifted scale).  
     - **"Enthalpy"**: The cumulative **heat** (in J, kJ, etc., depending on your $C_p$).

This **composite curve** is generally used alongside **pinch analysis** to visualize the **hot** and **cold** energy demands over the **temperature range**.


```python
def build_composite_curve(process_data: pd.DataFrame,
                          dT_min: float = 10.0,
                          stream_type: str = 'hot') -> pd.DataFrame:
    """
    Build a composite curve (hot or cold) as a DataFrame of (Temperature, Enthalpy),
    using the user-specified dT_min.

    For hot streams:
      - Only consider rows indexed "hot".
      - Shift them downward by dT_min/2.

    For cold streams:
      - Only consider rows indexed "cold".
      - Shift upward by dT_min/2.

    Returns a DataFrame with columns:
      - 'Temperature' (shifted temperature scale)
      - 'Enthalpy' (cumulative heat release/absorption as we move downward)
    """

    # 1. Filter the DataFrame by stream type
    if stream_type.lower() == 'hot':
        df = process_data.loc[[idx for idx in process_data.index if idx.lower() == 'hot']]
        shift_sign = -1  # shift down for hot
    else:
        df = process_data.loc[[idx for idx in process_data.index if idx.lower() == 'cold']]
        shift_sign = 1   # shift up for cold

    # 2. Create shifted DataFrame (copies the original subset)
    shifted_df = df.copy()

    # Shift Tsupply (¬∞C) and Ttarget (¬∞C) by +/- dT_min/2
    shifted_df['Tsupply_shifted'] = shifted_df.apply(
        lambda row: row['Tsupply (¬∞C)'] + shift_sign * (dT_min / 2), axis=1
    )
    shifted_df['Ttarget_shifted'] = shifted_df.apply(
        lambda row: row['Ttarget (¬∞C)'] + shift_sign * (dT_min / 2), axis=1
    )

    # 3. Collect all unique shifted temperatures and sort (descending)
    temps = []
    for idx, row in shifted_df.iterrows():
        temps.append(row['Tsupply_shifted'])
        temps.append(row['Ttarget_shifted'])
    unique_temps = sorted(set(temps), reverse=True)

    # 4. Build intervals between successive unique temperatures
    intervals = [
        (unique_temps[i], unique_temps[i+1])
        for i in range(len(unique_temps) - 1)
    ]

    # We'll construct the composite curve from the top to bottom
    enthalpy_list = [0.0]   # start from 0 at the top
    running_enthalpy = 0.0

    # 5. Step through each interval
    for (T_high, T_low) in intervals:
        deltaT = T_high - T_low
        # Sum (positive) Cp for streams spanning this entire interval
        Cp_net = 0.0
        for i2, row in shifted_df.iterrows():
            s_in = row['Tsupply_shifted']
            t_in = row['Ttarget_shifted']
            cp_val = row['Cp (J/(kg*K))']  # updated column name

            # If the interval (T_high -> T_low) is fully within s_in -> t_in
            if (max(s_in, t_in) >= T_high) and (min(s_in, t_in) <= T_low):
                # Add this stream's Cp to the net
                Cp_net += cp_val

        # Heat in this interval
        dQ = Cp_net * deltaT
        running_enthalpy += dQ
        enthalpy_list.append(running_enthalpy)

    # 6. Build DataFrame with boundary temperatures
    # We have (len(intervals)+1) boundary temps and enthalpy points
    curve_df = pd.DataFrame({
        'Temperature': unique_temps,
        'Enthalpy': enthalpy_list
    })

    return curve_df
```

# <span style="color:#2C3E50;">**plot_composite_curves with User-Selected Units**</span>

This enhanced version of `plot_composite_curves` lets you specify:

1. <span style="color:#2980B9;">**Temperature Units** (X-axis)</span>  
   - **'C'** for Celsius (default)  
   - **'K'** for Kelvin (adds **273.15** to all temperatures)

2. <span style="color:#2980B9;">**Enthalpy Units** (Y-axis)</span>  
   - Chosen via the **`q_unit_factors`** dictionary, e.g.:  
     - `'1'` ‚Üí J  
     - `'2'` ‚Üí kJ  
     - `'3'` ‚Üí MJ  
     - `'4'` ‚Üí kWh  
     - `'5'` ‚Üí cal  
     - `'6'` ‚Üí kcal  
     - `'7'` ‚Üí BTU  
     - `'8'` ‚Üí MWh  

---

## <span style="color:#16A085;">**Key Steps**</span>

1. **Build Composite Curves**  
   - Calls `build_composite_curve` for **hot** and **cold** streams in the **standard domain** (¬∞C, J).

2. **Grand Composite Curve (GCC)**  
   - Merges **hot** and **cold** curves on a **common temperature** grid,  
   - Interpolates missing points,  
   - Subtracts **cold** from **hot** to yield a **net enthalpy** (`hot - cold`).

3. **Temperature Scaling**  
   - If `temp_unit='K'`, adds **273.15** to the temperature axis for **all curves** (hot, cold, grand).

4. **Enthalpy Scaling**  
   - If `enthalpy_unit` ‚â† '1' (where '1' means J), we retrieve the relevant factor from **`q_unit_factors`**  
     and **invert** it (e.g., dividing by 1000 for kJ) to convert from J to the chosen unit.

5. **Plot**  
   - Displays **hot**, **cold**, and **grand** curves on a single **matplotlib** figure, each with distinct styles.  
   - Labels and legend reflect your chosen units.

6. **Return**  
   - Returns `(fig, ax)`, letting you **save** or further **customize** the plot.

---

**Why This Matters?**  
By **separating** internal (SI) data from **plot units**, you maintain **consistent** analysis (¬∞C, J) while giving a **user-friendly** plot in **C/K** and **J/kJ/MJ/...**. It‚Äôs the best of both worlds: accurate **analysis** plus **readable** visual output.


```python
def plot_composite_curves(
    process_data: pd.DataFrame,
    dT_min: float = 10.0,
    temp_unit: str = 'C',           # 'C' or 'K'
    enthalpy_unit: str = '1'        # '1' => J, '2' => kJ, etc., referencing your q_unit_factors
):
    """
    Plot the Hot Composite Curve, Cold Composite Curve, and Grand Composite Curve.
    Allows user-chosen units for temperature (C or K) and enthalpy (from q_unit_factors).

    Parameters
    ----------
    process_data : pd.DataFrame
        DataFrame with columns:
         - "Tsupply (¬∞C)"
         - "Ttarget (¬∞C)"
         - "Cp (J/(kg*K))"
        Rows labeled "hot" or "cold".

    dT_min : float
        Minimum temperature difference for shifting.

    temp_unit : str
        'C' for plotting temperature in ¬∞C, 'K' for Kelvin.

    enthalpy_unit : str
        A key into q_unit_factors dict:
          '1' => J
          '2' => kJ
          '3' => MJ
          etc.
        This will rescale the enthalpy axis.

    Returns
    -------
    (fig, ax): matplotlib figure and axes objects
    """

    # 1. Build the hot composite curve in the standard (¬∞C, J) scale
    hot_curve = build_composite_curve(process_data, dT_min=dT_min, stream_type='hot')
    # 2. Build the cold composite curve
    cold_curve = build_composite_curve(process_data, dT_min=dT_min, stream_type='cold')
    
    # --- Construct the Grand Composite Curve (GCC) numerically ---
    # Merge the two dataframes on a common temperature grid
    all_temps = sorted(
        set(hot_curve['Temperature'].tolist() + cold_curve['Temperature'].tolist()),
        reverse=True
    )
    
    hot_curve_interp = (hot_curve
                        .set_index('Temperature')
                        .reindex(all_temps)
                        .interpolate(method='index'))
    cold_curve_interp = (cold_curve
                         .set_index('Temperature')
                         .reindex(all_temps)
                         .interpolate(method='index'))
    
    gcc_values = hot_curve_interp['Enthalpy'] - cold_curve_interp['Enthalpy']
    grand_curve = pd.DataFrame({
        'Temperature': all_temps,
        'Enthalpy': gcc_values
    }).reset_index(drop=True)
    
    # --------------------------------------------------
    # 3. Convert Temperature if 'K' chosen
    # --------------------------------------------------
    if temp_unit.upper() == 'K':
        hot_curve['Temperature']   = hot_curve['Temperature']   + 273.15
        cold_curve['Temperature']  = cold_curve['Temperature']  + 273.15
        grand_curve['Temperature'] = grand_curve['Temperature'] + 273.15
        temperature_label = "Shifted Temperature (K)"
    else:
        temperature_label = "Shifted Temperature (¬∞C)"
    
    # --------------------------------------------------
    # 4. Convert Enthalpy if user selected something other than 'J'
    #    We'll use the q_unit_factors dict from your prior definitions
    # --------------------------------------------------
    # E.g. if enthalpy_unit == '2', that means kJ => we want to divide by 1000 if the data is in J
    # (or multiply by the factor's inverse).
    from math import isclose

    # We'll import or define the dictionary again here for clarity, or assume it's in scope:
    # q_unit_factors = {
    #     '1': 1.0, 
    #     '2': 1000.0,  # kJ => J
    #     '3': 1e6, ...
    # }
    # We invert the factor so that if the dictionary is "kJ => J" (factor=1000.0),
    # but our data is currently in J, we do data / 1000 to get kJ.
    
    # For example:
    # If enthalpy_unit='2' => factor=1000. Then actual scale factor for data is 1/factor = 1/1000
    # so that we end up in kJ on the Y-axis.
    
    # We'll define a small function to invert the factor:
    scale_factor = 1.0
    if enthalpy_unit in q_unit_factors:
        orig_factor = q_unit_factors[enthalpy_unit]  # e.g. 1000 for kJ => J
        # We want to invert it to go from J => kJ
        if not isclose(orig_factor, 0.0):
            scale_factor = 1.0 / orig_factor
    else:
        # default => 1.0 (no scaling)
        print("Warning: unrecognized enthalpy_unit. No conversion applied.")
    
    if scale_factor != 1.0:
        hot_curve['Enthalpy']   = hot_curve['Enthalpy']   * scale_factor
        cold_curve['Enthalpy']  = cold_curve['Enthalpy']  * scale_factor
        grand_curve['Enthalpy'] = grand_curve['Enthalpy'] * scale_factor
    
    # Determine label for the Y-axis
    # We'll do a simple mapping for demonstration. You can refine as needed.
    y_label_map = {
        '1': "Enthalpy (J)",
        '2': "Enthalpy (kJ)",
        '3': "Enthalpy (MJ)",
        '4': "Enthalpy (kWh?)",  # or whichever label you want
        '5': "Enthalpy (cal)",
        '6': "Enthalpy (kcal)",
        '7': "Enthalpy (BTU)",
        '8': "Enthalpy (MWh)"
    }
    y_label = y_label_map.get(enthalpy_unit, "Enthalpy (J)")
    
    # --------------------------------------------------
    # 5. Plot
    # --------------------------------------------------
    fig, ax = plt.subplots(figsize=(7, 5))

    ax.plot(hot_curve['Temperature'], hot_curve['Enthalpy'], 'r-o', label='Hot Composite Curve')
    ax.plot(cold_curve['Temperature'], cold_curve['Enthalpy'], 'b-o', label='Cold Composite Curve')
    ax.plot(grand_curve['Temperature'], grand_curve['Enthalpy'], 'g--', label='Grand Composite Curve')
    
    ax.set_xlabel(temperature_label)
    ax.set_ylabel(y_label)
    ax.set_title('Composite Curves & Grand Composite Curve')
    ax.grid(True)
    ax.legend()
    
    return fig, ax
```
