# Curve Fitting with the Michaelis-Menten Model

In enzyme kinetics, the **Michaelis-Menten model** describes the relationship between the substrate concentration and the reaction rate. This model is given by the equation:

$$
v = \frac{V_{\text{max}} \cdot [S]}{K_m + [S]}
$$

Where:
- The reaction rate is denoted as $v$.
- The substrate concentration is represented as $[S]$.
- The maximum reaction rate is denoted as $V_{\text{max}}$.
- The Michaelis constant is represented as $K_m$ (the substrate concentration at half-maximal velocity).

To fit the model to experimental data, we can use the `curve_fit` function from the `scipy.optimize` module. This function estimates the values of $V_{\text{max}}$ and $K_m$ that best fit the observed reaction rates based on the substrate concentrations.

Hereâ€™s a breakdown of how we perform the curve fitting:

1. **`curve_fit` function**: This function is used to fit a given model to experimental data. It minimizes the difference between the observed data and the predicted values from the model.

   - **Input parameters**:
     - `michaelis_menten`: The function we are fitting, which is the Michaelis-Menten equation:
       $$
       v = \frac{V_{\text{max}} \cdot [S]}{K_m + [S]}
       $$
     - `substrate_conc`: The independent variable (x-axis data), representing the substrate concentration.
     - `reaction_rates`: The dependent variable (y-axis data), representing the observed reaction rate.
     - `bounds`: We specify the bounds for the fitting parameters. In this case, we allow the parameters to be non-negative by setting the lower bound to 0 and allowing the upper bound to be infinity (no upper limit for the parameters).
     - 

2. **Output of `curve_fit`**: 
   - The function returns the optimal values for $V_{\text{max}}$ and $K_m$, which are stored in the variable `params`.

3. **Using the fitted parameters**: Once we have the fitted values of $V_{\text{max}}$ and $K_m$, we can generate the fitted curve. This is done by calculating the reaction rate at various substrate concentrations using the fitted parameters.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Michaelis-Menten function
def michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)
```

```python
# Load data from CSV
file_path = 'mich-men-random_csv_data.csv'  # Example experimental data CSV file path
data = pd.read_csv(file_path)
```

```python
# Extract substrate concentration and reaction rates
substrate_conc = data['substrate_conc'].values
reaction_rates = data['reaction_rates'].values
```

```python
# Curve fitting
params, _ = curve_fit(michaelis_menten, substrate_conc, reaction_rates, bounds=(0, [np.inf, np.inf]))
Vmax, Km = params

# Generate fitted curve
S_fit = np.linspace(0, max(substrate_conc), 100)
v_fit = michaelis_menten(S_fit, Vmax, Km)
```

```python
from datetime import datetime

# Create a figure and axis
plt.figure(figsize=(10, 6))

# Plot experimental data and fitted curve
plt.scatter(substrate_conc, reaction_rates, label="Experimental Data", color="blue", s=50, edgecolor='black')
plt.plot(S_fit, v_fit, label=f"Fitted Curve (Vmax={Vmax:.2f}, Km={Km:.2f})", color="red", linewidth=2)

# Add labels and title
plt.xlabel("Substrate Concentration ([S])", fontsize=14)
plt.ylabel("Reaction Rate (v)", fontsize=14)
plt.title("Michaelis-Menten Kinetics", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Generate the current date and time for the filename
current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"mich-men-plot-{current_time}.jpg"

# Save the plot as a JPG file
plt.savefig(filename, format='jpg', dpi=300)

# Show the plot
plt.show()

# Output the results
print(f"Vmax: {Vmax:.2f}")
print(f"Km: {Km:.2f}")
```

```python
import ipywidgets as widgets
import cv2
import numpy as np
import matplotlib.pyplot as plt
import csv

from IPython.display import display
from google.colab import files  # For downloading the CSV in Colab

# Ensure inline plotting
%matplotlib inline

# ----------------- WIDGETS -----------------
uploader = widgets.FileUpload(accept='.png,.jpg,.jpeg', multiple=False)
slider = widgets.IntSlider(value=5, min=1, max=20, step=1, description='Line Pick')
extract_button = widgets.Button(description="Extract & Save Data")
output_label = widgets.Label("Upload an image and click to extract.")
ui = widgets.VBox([output_label, uploader, slider, extract_button])
display(ui)

def save_data_to_csv(data_points, filename="extracted_data.csv"):
    """Save a list of (x, y) points to a CSV file."""
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["x_pixel", "y_pixel"])  # Header
        writer.writerows(data_points)
    print(f"Data saved to {filename}")

def extract_curve_points(edges, strategy='median'):
    """
    Scans each x-column in 'edges' and extracts a single y-coordinate
    based on the chosen strategy:
      - 'top': pick the topmost edge pixel
      - 'bottom': pick the bottommost edge pixel
      - 'median': pick the median of edge pixels in that column
    Returns a list of (x, y) points.
    """
    height, width = edges.shape
    extracted_points = []

    for x in range(width):
        # Find all y indices in this column that are edges (==255)
        y_indices = np.where(edges[:, x] == 255)[0]
        if len(y_indices) > 0:
            if strategy == 'top':
                y = y_indices[0]
            elif strategy == 'bottom':
                y = y_indices[-1]
            else:  # 'median'
                y = int(np.median(y_indices))
            extracted_points.append((x, y))
    
    return extracted_points

def on_extract_button_click(_):
    # 1) Check if a file was uploaded
    if not uploader.value:
        output_label.value = "Please upload an image file!"
        return
    
    # 2) Grab the uploaded image from the widget
    file_data = list(uploader.value.values())[0]['content']  # first (only) file
    np_arr = np.frombuffer(file_data, np.uint8)
    img = cv2.imdecode(np_arr, cv2.IMREAD_GRAYSCALE)
    if img is None:
        output_label.value = "Error decoding image. Try another file."
        return
    
    # 3) Threshold (Otsu) and edge-detect
    #    We invert the image first if the background is dark and the line is lighter.
    inverted = 255 - img
    _, thresh = cv2.threshold(inverted, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    edges = cv2.Canny(thresh, 50, 150)

    # 4) Extract curve points from edges
    #    The slider can be used to choose one of three strategies:
    #      1-7: 'top'    8-14: 'median'    15-20: 'bottom'
    val = slider.value
    if val <= 7:
        strategy = 'top'
    elif val <= 14:
        strategy = 'median'
    else:
        strategy = 'bottom'
    
    extracted_points = extract_curve_points(edges, strategy=strategy)
    output_label.value = f"Extracted {len(extracted_points)} points using '{strategy}' strategy."

    # 5) Plot the results
    plt.figure(figsize=(8, 6))
    plt.imshow(img, cmap='gray')
    xs, ys = zip(*extracted_points) if extracted_points else ([], [])
    plt.scatter(xs, ys, color='red', label='Extracted Points')
    plt.legend()
    plt.title("Extracted Data Points")
    plt.axis("off")
    plt.show()
    
    # 6) Save to CSV in the working directory
    save_data_to_csv(extracted_points, filename="extracted_data.csv")
    
    # 7) Download the CSV automatically in Colab
    files.download("extracted_data.csv")

# 8) Connect the button to the callback
extract_button.on_click(on_extract_button_click)
```
